```python
import dlt
from pyspark.sql.functions import current_timestamp, col, lit


# CUSTOMERS

@dlt.table(
    name="customers_bronze",
    comment="Raw customers data",
    table_properties={"quality": "bronze"}
)
@dlt.expect("valid_customer_id", "customer_id IS NOT NULL")
def customers_bronze():
    return (
        spark.readStream.format("cloudFiles")
        .option("cloudFiles.format", "csv")
        .option("header", "true")
        .option("cloudFiles.inferColumnTypes", "true")
        .option("cloudFiles.includeExistingFiles", "true")
        .option("cloudFiles.schemaEvolutionMode", "addNewColumns")
        .option("cloudFiles.schemaLocation", "s3://storereailssales/schema/customers")
        .load("s3://storereailssales/raw/customers/")
        .withColumn("source_system", lit("CRM"))
        .withColumn("ingestion_ts", current_timestamp())
        .withColumn("source_file", col("_metadata.file_path"))
    )


# PRODUCTS

@dlt.table(
    name="products_bronze",
    comment="Raw products data",
    table_properties={"quality": "bronze"}
)
@dlt.expect("valid_product_id", "product_id IS NOT NULL")
def products_bronze():
    return (
        spark.readStream.format("cloudFiles")
        .option("cloudFiles.format", "csv")
        .option("header", "true")
        .option("cloudFiles.inferColumnTypes", "true")
        .option("cloudFiles.includeExistingFiles", "true")
        .option("cloudFiles.schemaEvolutionMode", "addNewColumns")
        .option("cloudFiles.schemaLocation", "s3://storereailssales/schema/products")
        .load("s3://storereailssales/raw/products/")
        .withColumn("source_system", lit("ERP"))
        .withColumn("ingestion_ts", current_timestamp())
        .withColumn("source_file", col("_metadata.file_path"))
    )


# SALES

@dlt.table(
    name="sales_bronze",
    comment="Raw sales data",
    table_properties={"quality": "bronze"}
)
@dlt.expect("valid_order_id", "order_id IS NOT NULL")
def sales_bronze():
    return (
        spark.readStream.format("cloudFiles")
        .option("cloudFiles.format", "csv")
        .option("header", "true")
        .option("cloudFiles.inferColumnTypes", "true")
        .option("cloudFiles.includeExistingFiles", "true")
        .option("cloudFiles.schemaEvolutionMode", "addNewColumns")
        .option("cloudFiles.schemaLocation", "s3://storereailssales/schema/sales")
        .load("s3://storereailssales/raw/sales/")
        .withColumn("source_system", lit("POS"))
        .withColumn("ingestion_ts", current_timestamp())
        .withColumn("source_file", col("_metadata.file_path"))
    )



# STORES

@dlt.table(
    name="stores_bronze",
    comment="Raw stores data",
    table_properties={"quality": "bronze"}
)
@dlt.expect("valid_store_id", "store_id IS NOT NULL")
def stores_bronze():
    return (
        spark.readStream.format("cloudFiles")
        .option("cloudFiles.format", "csv")
        .option("header", "true")
        .option("cloudFiles.inferColumnTypes", "true")
        .option("cloudFiles.includeExistingFiles", "true")
        .option("cloudFiles.schemaEvolutionMode", "addNewColumns")
        .option("cloudFiles.schemaLocation", "s3://storereailssales/schema/stores")
        .load("s3://storereailssales/raw/stores/")
        .withColumn("source_system", lit("POS"))
        .withColumn("ingestion_ts", current_timestamp())
        .withColumn("source_file", col("_metadata.file_path"))
    )



# INVENTORY EVENTS

@dlt.table(
    name="inventory_events_bronze",
    comment="Raw inventory events data",
    table_properties={"quality": "bronze"}
)
def inventory_events_bronze():
    return (
        spark.readStream.format("cloudFiles")
        .option("cloudFiles.format", "csv")
        .option("header", "true")
        .option("cloudFiles.inferColumnTypes", "true")
        .option("cloudFiles.includeExistingFiles", "true")
        .option("cloudFiles.schemaEvolutionMode", "addNewColumns")
        .option("cloudFiles.schemaLocation", "s3://storereailssales/schema/inventory_events")
        .load("s3://storereailssales/raw/inventory_events/") 
        .withColumn("source_system", lit("IoT"))
        .withColumn("ingestion_ts", current_timestamp())
        .withColumn("source_file", col("_metadata.file_path"))
    )
```
