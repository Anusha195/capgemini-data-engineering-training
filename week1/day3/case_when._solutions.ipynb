```sql
-- salary hike based on performance
select * ,
case 
  when experience >=8 and performance_rating = 'A' then salary*1.20 
  when experience >= 5 and performance_rating = 'B' then salary * 1.15
  when performance_rating = 'C' then salary
  else salary*1.10
end as salary_hike from Employee_data
```

```sql
SELECT emp_id, emp_name, salary, performance_rating,
       CASE
           WHEN salary > 80000 AND performance_rating = 'A' THEN 'High Performer'
           WHEN salary BETWEEN 50000 AND 80000 AND performance_rating = 'B' THEN 'Mid Performer'
           WHEN salary < 50000 OR performance_rating = 'C' THEN 'Low Performer'
           ELSE 'Others'
       END AS category
FROM Employee_data;
```

```sql
SELECT 
    emp_id,
    emp_name,
    salary,
    experience,

    CASE 
        WHEN salary > 90000 AND experience > 10 THEN '35%'
        WHEN salary > 90000 AND experience <= 10 THEN '30%'
        
        WHEN salary BETWEEN 60000 AND 90000 AND experience > 5 THEN '25%'
        WHEN salary BETWEEN 60000 AND 90000 AND experience <= 5 THEN '20%'
        
        WHEN salary < 60000 THEN '15%'
        
    END AS tax_rate

FROM employee_data;
```

```sql
SELECT 
    emp_id,
    emp_name,
    salary,
    performance_rating,
    experience,
    
    CASE 
        WHEN performance_rating = 'A' THEN
            CASE 
                WHEN salary > 80000 AND experience > 5 THEN salary * 1.25
                WHEN salary BETWEEN 50000 AND 80000 THEN salary * 1.15
                ELSE salary * 1.20
            END
        
        WHEN performance_rating = 'B' THEN
            CASE 
                WHEN experience > 5 THEN salary * 1.12
                ELSE salary * 1.10
            END
        
        WHEN performance_rating = 'C' THEN salary
        
    END AS new_salary

FROM employee_data;
```

```sql
SELECT 
    emp_id,
    emp_name,
    department,
    salary,
    performance_rating,
    experience,

    CASE 
        WHEN department = 'Finance' THEN
            CASE 
                WHEN performance_rating = 'A' AND experience > 10 THEN salary * 0.25
                ELSE salary * 0.20
            END

        WHEN department = 'HR' THEN
            CASE 
                WHEN performance_rating = 'B' OR experience > 5 THEN salary * 0.15
                ELSE salary * 0.10
            END

        ELSE salary * 0.08
    END AS bonus

FROM employee_data;
```
