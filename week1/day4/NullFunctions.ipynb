Employees with NULL salary

```sql
SELECT * 
FROM Employees4
WHERE salary IS NULL;
```

Orders where discount is NOT NULL

```sql
SELECT * 
FROM Orders4
WHERE discount IS NOT NULL;
```

Products where category is NULL

```sql
SELECT * 
FROM Products4
WHERE category IS NULL;
```

Count employees with NULL manager_id

```sql
SELECT COUNT(*) AS count_null_manager
FROM Employees4
WHERE manager_id IS NULL;
```

Replace NULL salary with 0

```sql
SELECT emp_id, IFNULL(salary, 0) AS salary
FROM Employees4;
```

Replace NULL bonus with 1000

```sql
SELECT emp_id, IFNULL(bonus, 1000) AS bonus
FROM Employees4;
```

Replace NULL order amount with 500

```sql
SELECT order_id, IFNULL(amount, 500) AS amount
FROM Orders4;
```

Replace NULL stock with 0

```sql
SELECT product_id, IFNULL(stock, 0) AS stock
FROM Products4;
```

Employee earnings (salary else bonus) using coalesce

```sql
SELECT emp_id, COALESCE(salary, bonus) AS earnings
FROM Employees4;
```

salary → bonus → 0

```sql
SELECT emp_id, COALESCE(salary, bonus, 0) AS income
FROM Employees4;
```

Salary = 0 → NULL using NULLIF

```sql
SELECT emp_id, NULLIF(salary, 0) AS salary
FROM Employees4;
```

Employee income (fallback chain)

```sql
SELECT emp_id,
       COALESCE(salary, bonus, 1000) AS income
FROM Employees4;
```

Avoid divide by zero

```sql
SELECT amount / NULLIF(discount, 0)
FROM Orders4;
```

Final payable amount

```sql
SELECT order_id,
       IFNULL(amount, 0) - IFNULL(discount, 0) AS final_amount
FROM Orders4;
```
